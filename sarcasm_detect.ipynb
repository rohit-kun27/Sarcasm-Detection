```python
import pandas as pd
```

```python
df = pd.read_json("Sarcasm_Headlines_Dataset.json", lines=True)
```

```python
df.head()
```

```python
df
```

```python
df = df[['headline', 'is_sarcastic']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')

# Remove stopwords and split dataset
stop_words = set(stopwords.words('english'))
df['headline'] = df['headline'].str.lower().str.replace('[^\w\s]', '')
df['headline'] = df['headline'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))

X_train, X_test, y_train, y_test = train_test_split(df['headline'], df['is_sarcastic'], test_size=0.2, random_state=42)

# Convert text to TF-IDF features
vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

# Train logistic regression model
model = LogisticRegression()
model.fit(X_train_tfidf, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score

# Make predictions
y_pred = model.predict(X_test_tfidf)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import joblib

# Save the model
joblib.dump(model, "sarcasm_model.pkl")

# Save the vectorizer
joblib.dump(vectorizer, "tfidf_vectorizer.pkl")
```
